```python
import re
def ex_num(s):
    num = re.findall(r'\d+', s)
    return(int(num[0]))
```

```python
orig = "is it known that Hong kong is the city where almog is at"
clus = "is it known that (0, 0) is the (0) where (1) is at"
res = ""

def new_format(orig,clus):
    res = ""
    clus_ls = clus.split(" ")
    orig_ls = orig.split(" ")
    to = min(len(orig.split(" ")), len(clus.split(" ")))
    for i in range(to):
        in_clus = 0
        if len(clus_ls[i])>0:
            if clus_ls[i][0] == "(":
                res = res + " <start_{0}> ".format(ex_num(clus_ls[i])) + orig_ls[i]
                in_clus = 1
            if clus_ls[i][-1] == ")":
                if in_clus == 1:
                    res = res + " <end_{0}>".format(ex_num(clus_ls[i]))
                else:
                    res = res + " " + orig_ls[i] + " <end_{0}>".format(ex_num(clus_ls[i]))
                in_clus = 1
            if in_clus == 0:
                res = res + " " + orig_ls[i]
    if(len(res)>1):
        if res[0] == " ":
            res = res[1:]
        if res[-1] == " ":
            res = res[:-1]
    return res
```

```python
def is_start(f):
    if f[0] in ("<", "(", "[") and f[-1] in (">", ")", "]"):
        if not any(i.isdigit() for i in f):
            return False
        if "endels" in f:
            return False
        if "start" in f or "s" in f or "begin" in f:
            return True
    return False


def is_end(f):
    print(f)
    if any(substring in f for substring in ["<", "(", "["]) and any(substring in f for substring in [">", ")", "]"]):
        if "end" in f:
            return True
    return False
```
